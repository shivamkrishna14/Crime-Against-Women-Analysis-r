```python
#Year vs Rape
fig = plt.figure(figsize=(10,5))
sns.barplot(x='Year',y='Rape',data=crimedata,ci=0)
plt.title('N0.of Rape cases in Year 2001-2014')
plt.xlabel('Year')
plt.ylabel('Rape_Cases')
plt.show()
```

```python
state_wise = crimedata.pivot_table(values='Rape',index = 'STATE/UT',aggfunc='sum').reset_index()


plt.figure(figsize=(10, 15))

plt.style.use('ggplot')
sns.barplot(x = 'Rape', y = 'STATE/UT', data = state_wise,edgecolor = None)

plt.title("Rapes in each state",fontdict={'fontsize':30},pad = 30, color = 'black')
```

As we can see here that the, Madhay Pradesh has the highest number of rape cases followed by Maharastra, Uttar Pradesh, West Bengal and Rajasthan.

```python
y_o_y = crimedata.pivot_table(values='Rape',index='Year',aggfunc='sum').reset_index()


plt.figure(figsize=(15, 8))
plt.style.use('ggplot')
sns.lineplot(x = 'Year', y = 'Rape' , data = y_o_y, color = 'blue')
plt.title("Rape Victims Year on Year",fontdict={'fontsize':30},pad = 30,color = 'black')
```

```python
state_wise = crimedata.pivot_table(values='Kidnapping and Abduction',index = 'STATE/UT',aggfunc='sum').reset_index()


plt.figure(figsize=(10, 15))

plt.style.use('ggplot')
sns.barplot(x = 'Kidnapping and Abduction', y = 'STATE/UT', data = state_wise,edgecolor = None)

plt.title("Kidnapping and Abduction in each state",fontdict={'fontsize':30},pad = 30, color = 'black')
```

As we can see here that, Uttar Pradesh has the highest number of kidnapping and abduction cases followed by Rajasthan and West Bengal.

```python
state_wise = crimedata.pivot_table(values='Dowry Deaths',index = 'STATE/UT',aggfunc='sum').reset_index()


plt.figure(figsize=(10, 15))

plt.style.use('ggplot')
sns.barplot(x = 'Dowry Deaths', y = 'STATE/UT', data = state_wise,edgecolor = None)

plt.title("Dowry Deaths Cases in each state",fontdict={'fontsize':30},pad = 30, color = 'black')
```

```python
state_wise = crimedata.pivot_table(values='Importation of Girls',index = 'STATE/UT',aggfunc='sum').reset_index()


plt.figure(figsize=(10, 15))

plt.style.use('ggplot')
sns.barplot(x = 'Importation of Girls', y = 'STATE/UT', data = state_wise,edgecolor = None)

plt.title("Importation of Girls Cases in each state",fontdict={'fontsize':30},pad = 30, color = 'black')
```
